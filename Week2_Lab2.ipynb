```python
import cv2
import matplotlib.pyplot as plt

image = cv2.imread('images/CAU.png')
image_RGB = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

add_image = cv2.add(image_RGB, 100)
subtract_image = cv2.subtract(image_RGB, 50)
multiply_image = cv2.multiply(image_RGB, 2)
divide_image = cv2.divide(image_RGB, 3.0)

plt.imshow(image_RGB)
plt.title('Original Image')
plt.axis('off')

plt.figure(figsize=(8,8))
plt.subplot(2, 2, 1)
plt.title('Add Image')
plt.imshow(add_image)
plt.axis('off')
plt.subplot(2, 2, 2)
plt.title('Subtract Image')
plt.imshow(subtract_image)
plt.axis('off')
plt.subplot(2, 2, 3)
plt.title('Multiply Image')
plt.imshow(multiply_image)
plt.axis('off')
plt.subplot(2, 2, 4)
plt.title('Divivde Image')
plt.imshow(divide_image)
plt.axis('off')
plt.show()
```
